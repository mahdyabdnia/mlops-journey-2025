```python
# ۲. pydantic (برای اعتبارسنجی داده)
!pip install pydantic -q

from pydantic import BaseModel
from typing import Literal

class User(BaseModel):
    name: str
    age: int
    role: Literal["admin", "user", "guest"]

# درست
u1 = User(name="مهدی", age=30, role="admin")
print(u1)

# اشتباه → خودش خطا می‌ده
# u2 = User(name="علی", age="سی", role="hacker")
```

```python
# ۳. dataclass
from dataclasses import dataclass

@dataclass
class Product:
    name: str
    price: float
    in_stock: bool = True

p = Product("آیفون ۱۶", 85_000_000, False)
print(p)
```
